```python
score = [90, 80, 88, 84, 85, 87, 88, 89, 90, 100]
score[0] = score[0]+10
score[9] = score[9]-10
print('更新後分數', score)
```

```python
score = []
print('目前串列內容 = ',score)
score.append('80')
print('目前串列內容 = ',score)
score.append('70')
print('目前串列內容 = ',score)
score.append('60')
print('目前串列內容 = ',score)
```

```python
score = [80, 90, 85, 75, 70, 65, 88, 98, 100]
score[3] = 78
print('最高分', max (score))
print('最低分', min (score))
print('總和', sum (score))
score.append(50)
score.append(99)
print(score)
score.sort
print('由小到大印出分數', sorted(score))
score.sort(reverse=True)
print('由大到小印出分數', sorted(score,reverse=True))
```
